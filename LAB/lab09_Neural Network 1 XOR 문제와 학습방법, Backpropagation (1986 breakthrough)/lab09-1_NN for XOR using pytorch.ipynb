```python
import torch
from torch.autograd import Variable
from torch import nn, optim
import numpy as np
import torch.nn.functional as F
```

```python
x_data = np.array([[0,0],[0,1],[1,0],[1,1]], dtype = np.float32)
y_data = np.array([[0],[1],[1],[0]], dtype = np.float32)
```

### 만약에, Logistic Regression 으로 XOR을 해보려고 하면??

```python
X = Variable(torch.Tensor(x_data))
Y = Variable(torch.Tensor(y_data))


class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear = torch.nn.Linear(2, 1,bias = True) # 2 in and 1 out
        
    def forward(self, x):
        y_pred = torch.sigmoid(self.linear(x))
        return y_pred


model = Model()
criterion = torch.nn.BCELoss(reduction='sum')
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
for t in range(10001):
    y_pred = model(X)
    cost = criterion(y_pred,Y)
    
    optimizer.zero_grad()
    cost.backward(retain_graph=True)
    optimizer.step()
    if t % 20 == 0:
        print(t, "\ncost : ",cost.data.numpy(), "\nweight : \n",model.linear.weight.data.numpy(), model.linear.bias.data.numpy())
```

```python

predicted = (model(X).data > 0.5).float()
accuracy = (predicted == Y.data).float().mean()
print("Hypotheis : \n",y_pred.detach().numpy(), \
      "\nCorrect : \n",predicted.detach().numpy(),\
      "\nAccuracy : ",accuracy.numpy())
```

학습이 잘 안된다! 아무리 많이 돌려도, 0.5를 넘을 수 없다. -> 이론적으로, XOR은 logistic Regression 으로 불가능하다.

### Neural Net을 이용한다면?

```python
x_data = np.array([[0,0],[0,1],[1,0],[1,1]], dtype = np.float32)
y_data = np.array([[0],[1],[1],[0]], dtype = np.float32)
X = Variable(torch.Tensor(x_data))
Y = Variable(torch.Tensor(y_data))
```

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.l1 = torch.nn.Linear(2, 2,bias = True)
        self.l2 = torch.nn.Linear(2, 1,bias = True)
    def forward(self, x):
        y1 = torch.sigmoid(self.l1(x))
        y2 = torch.sigmoid(self.l2(y1))
        return y2


model = Model()
criterion = torch.nn.BCELoss(reduction='sum')
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

```python
for t in range(10001):
    y_pred = model(X)
    cost = criterion(y_pred,Y)
    
    optimizer.zero_grad()
    cost.backward(retain_graph=True)
    optimizer.step()
    if t % 20 == 0:
        print(t, "\ncost : ",cost.data.numpy())
```

```python

predicted = (model(X).data > 0.5).float()
accuracy = (predicted == Y.data).float().mean()
print("Hypotheis : \n",y_pred.detach().numpy(), \
      "\nCorrect : \n",predicted.detach().numpy(),\
      "\nAccuracy : ",accuracy.numpy())
```
