```python
import torch
from torch.autograd import Variable
from torch import nn, optim
import numpy as np
import torch.nn.functional as F
```

```python
# Predicting animal type based based on various features
xy = np.loadtxt('data-04-zoo.csv',delimiter = ',', dtype = np.float)
x_data = xy[:,0:-1]
y_data = xy[:,[-1]]
X = Variable(torch.from_numpy(x_data))
Y = Variable(torch.from_numpy(y_data))
nb_classes = 7 # 0 ~6의 숫자.
```

[make_one_hot](https://gist.github.com/jacobkimmel/4ccdc682a45662e514997f724297f39f)

```python
def make_one_hot(labels,nb_classes):
    one_hot = torch.FloatTensor(labels.size()[0],nb_classes).zero_()
    target = one_hot.scatter_(1, labels.long().data, 1)
    target = Variable(target)
    return target
Y_one_hot = make_one_hot(Y,nb_classes)

```

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear = nn.Linear(16, 7,bias = True) # 2 in and 1 out
        
    def forward(self, x):
        x = x.view(-1, 16)
        y_pred = self.linear(x)
        F.softmax(y_pred)
        return y_pred
    
model = Model()
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

```python
for t in range(10001):    
    optimizer.zero_grad()

    y_pred = model(X.float())
    cost = criterion(y_pred,Y.long().view(-1))
    cost.backward()

    optimizer.step()
    
    prediction = y_pred.argmax(dim=1).float()[1]
    correct_prediction = (prediction.data == Y.data.float())
    accuracy = correct_prediction.float().mean()
    
    if t % 200 == 0:
        print("step: ",t, "\ncost : ",cost.data.numpy())
        
        
```

```python
pred = y_pred.argmax(dim=1)
for p, y in zip(pred, Y):
    print(" Prediction: ",p.data.numpy(),"True Y:",y.data.numpy()[0])
```
