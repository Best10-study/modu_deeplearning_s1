# LAB03_Linear Regression using pytorch

## Hypothesis and cost function
$$H(x) = Wx + b$$
$$ cost(W,b) = \frac{1}{m}\sum_{i=1}^{m}(H(x^{(i)})-y^{(i)})^2 $$

```python
import torch
```

$H(x) = Wx+b$

```python
# X and Y data
x_train = torch.tensor([1,2,3],dtype = torch.float)
y_train = torch.tensor([1,2,3],dtype = torch.float)

W = torch.randn(1, device=torch.device("cpu"))
b = torch.randn(1, device=torch.device("cpu"))
```

```python
# learning rate
learning_rate = 0.01
```

## Gradient desecent
$$ cost(W) = \frac{1}{m}\sum^{m}_{i=1}(Wx^{(i)}-y^{(i)})^2$$
$$ W:=W - \alpha\frac{1}{m}\sum^{m}_{i=1}(Wx^{(i)}-y^{(i)})x^{(i)}$$
$$ b:=b - \alpha\frac{1}{m}\sum^{m}_{i=1}(Wx^{(i)}-y^{(i)})$$

```python
for t in range(10001):
    # Our hypothesis XW+b
    hypothesis = x_train * W + b

    # # cost/Loss function
    cost = (hypothesis - y_train).pow(2).mean().item()
    gradient = ((hypothesis - y_train)* x_train).mean().item()
    b_gradient = (hypothesis - y_train).mean().item()
    W = W - learning_rate * gradient
    b = b - learning_rate * b_gradient
    if t % 20 == 0:
        print(t, cost, [W.item()], [b.item()])
```

## Plot model

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10,0.1)
y = W.item()*x + b.item()
```

```python
plt.plot(x,y)
plt.axis([0, 10, 0, 10])
plt.show()
```
