# LAB02_Linear Regression Using PyTorch

## Hypothesis and cost function
$$H(x) = Wx + b$$
$$ cost(W,b) = \frac{1}{m}\sum_{i=1}^{m}(H(x^{(i)})-y^{(i)})^2 $$

```python
import torch
import numpy as np
from torch import nn, optim
from torch.autograd import Variable
```

$H(x) = Wx+b$

$cost(W,b) = \frac{1}{m}\sum_{i=1}^{m}(H(x^{(i)})-y^{(i)})^2$

```python
# X and Y data
x_train = Variable(torch.tensor([[1],[2],[3]],dtype = torch.float))
y_train = Variable(torch.tensor([[1],[2],[3]],dtype = torch.float))


model = nn.Linear(1,1,bias=True)
criterion = torch.nn.MSELoss(size_average=False)
optimizer = torch.optim.SGD(model.parameters(), lr=0.001)
```

```python
for t in range(20001):
    y_pred = model(x_train)
    cost = criterion(y_pred, y_train)
    
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    if t % 20 == 0:
        print(t,'cost: ',cost.data.numpy(),model.weight.data.numpy(),model.bias.data.numpy())

# After training
hour_var = Variable(torch.Tensor([[4.0]]))
y_pred = model(hour_var)
print("predict : ",  4, model(hour_var).data[0][0])
```
